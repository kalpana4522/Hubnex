# App size and Installs.

```python
#Let us firstly create a function to segregate apps base don their size
def app_classifier(size):
  '''This app takes the size of the app as the input and classifies it into small, medium and large size apps'''
  if size>0 and size<25000:   #size less than 25MB
    return 'Small App'
  elif size>=25000 and size<50000:   #Size less than 50MB
    return 'Medium App'
  elif size>=50000:     #size more than 50MB
    return 'Large App'
  else:
    return 'Invalid Size'
  return size
```

```python
#Lets apply the above function using lambda on Size column to classify apps
app_data['Size_type']= app_data['Size'].apply(lambda x: app_classifier(x))
app_data.head(5)
```

```python
#Lets now plot the app list as per the classification based on the size
plt.rcParams['figure.figsize']=(15,10)
app_data.groupby('Size_type')[['Installs']].sum().plot(kind='bar',fontsize=15)
plt.ylabel('Installs',fontsize=15)
plt.xlabel('Size Type',fontsize=15)
plt.title('Number of Installs based on Size',fontsize=15)
```

# App Size and Rating

```python
#size and rating
app_data.groupby('Size_type')[['Rating']].mean().plot(kind='bar')
plt.ylabel('Average Rating')
plt.title('Average Rating based on the App Size')
```
