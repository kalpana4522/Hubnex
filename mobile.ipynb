```python
import requests as rq
import pandas as pd
from bs4 import BeautifulSoup
```

```python
for i in range(2,10):
    url = 'https://www.flipkart.com/search?q=mobile&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'+str(i)
```

```python
Product_name = []
Prices = []
Descriptions = []
Ratings = []
```

```python
resp = rq.get(url)
# print(resp)
soup = BeautifulSoup(resp.text , 'html.parser')
box = soup.find("div" , class_ = "DOjaWF YJG4Cf")
# print(soup)
```

```python
box1 = soup.findAll("div" , class_ = "DOjaWF YJG4Cf")
box1
```

```python
box1
```

```python
names = box.find_all("div" , class_ = "KzDlHZ")
print(names)
```

```python
for i in names:
    name = i.text
    Product_name.append(name)
    # print(Product_name)

    # print(len(Product_name))
```

```python
prices = box.find_all("div" , class_ = "Nx9bqj _4b5DiR")
print(prices)
```

```python
for i in prices:
    name = i.text
    Prices.append(name)
    # print(Prices)
```

```python
desc = box.find_all("ul" , class_ = "G4BRas")
print(desc)
```

```python
for i in desc:
    name = i.text
    Descriptions.append(name)
    # print(Descriptions)
```

```python
reviews = box.find_all("div" , class_ = "XQDdHH")
print(reviews)
```

```python
for i in reviews:
    name = i.text
    Ratings.append(name)
    # print(Ratings)
```

```python
len(Ratings)
```

```python
df  =  pd.DataFrame({"Product_name" : Product_name, "Prices": prices, "Descriptions":Descriptions, "Ratings": Ratings})
print(df)
```

```python
df.to_csv("Flipkart_Data.csv")
```
